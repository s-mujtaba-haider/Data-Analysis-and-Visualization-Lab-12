```python
import re

def word_extraction(sentence):
    # This function should split the sentence while removing common stopwords using nltk.
    stopwords_list = set(stopwords.words('english'))
    words = re.findall(r'\b\w+\b', sentence.lower())
    cleaned_text = [word for word in words if word not in stopwords_list]
    return cleaned_text

def tokenize(sentences):
    # This function should tokenize a list of sentences using nltk.
    words = []
    for sentence in sentences:
        words.extend(word_extraction(sentence))
    return words

def calculateBOW(wordset, l_doc):
    # This function should calculate vectors (BOW representation) for each sentence in l_doc.
    vec = []
    for doc in l_doc:
        bow_vector = [0] * len(wordset)
        words = word_extraction(doc)
        for word in words:
            if word in wordset:
                bow_vector[wordset.index(word)] += 1
        vec.append(bow_vector)
    return vec

# Example usage:
sentences = ["This is a sample sentence.", "Another example sentence.", "A third sentence for testing."]
words = tokenize(sentences)
unique_words = list(set(words))
bow = calculateBOW(unique_words, sentences)

print("Unique words:", unique_words)
print("BOW representation:")
for i in range(len(sentences)):
    print(sentences[i], ":", bow[i])
```

# **QUESTION 02** : [12 Marks]

1. Create two filters for edge detection hx and hy of size 3x3. 'Use SOBEL FILTER'

2. for every 3*3 window of the image, multiply the two filters each at once, store the new result for the central pixel in two new images, one for the x value and one for the y value.

3. create a new image and fill it with the final value for edge from x and y filters together according to the following equation ( 𝑒𝑑𝑔𝑒𝑠𝑡𝑟𝑒𝑛𝑔𝑡ℎ = √(𝑥2 + 𝑦2 )^2 )

4. Apply threshold to the edge strength, in which edges above this threshold will be visible and others will disappear.
5. Print the image resulted from the (x filter only), (y filter only) and (from x and y filter) together.




```python
### Implementation
import cv2
import numpy as np

# Load your image
image = cv2.imread('your_image.jpg', cv2.IMREAD_GRAYSCALE)

# Define Sobel filters
hx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])  # Sobel filter for detecting edges in the x-direction
hy = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])  # Sobel filter for detecting edges in the y-direction

# Apply Sobel filters to the image
edge_x = cv2.filter2D(image, -1, hx)  # Convolve the image with the x-direction Sobel filter
edge_y = cv2.filter2D(image, -1, hy)  # Convolve the image with the y-direction Sobel filter

# Calculate edge strength
edge_strength = np.sqrt(edge_x ** 2 + edge_y ** 2)

# Apply a threshold to the edge strength
threshold = 50  # Adjust this threshold as needed
edge_strength[edge_strength < threshold] = 0  # Remove weak edges

# Display the results
cv2.imshow('Edge Detection (x filter)', edge_x)
cv2.imshow('Edge Detection (y filter)', edge_y)
cv2.imshow('Edge Detection (Combined)', edge_strength)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load your image
image = cv2.imread('your_image.jpg', cv2.IMREAD_GRAYSCALE)

# Define Laplacian filters
laplacian_x = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])  # Laplacian filter for detecting edges in the x-direction
laplacian_y = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])  # Laplacian filter for detecting edges in the y-direction

# Apply Laplacian filters to the image
edge_x = cv2.filter2D(image, -1, laplacian_x)  # Convolve the image with the x-direction Laplacian filter
edge_y = cv2.filter2D(image, -1, laplacian_y)  # Convolve the image with the y-direction Laplacian filter

# Calculate edge strength
edge_strength = np.sqrt(edge_x ** 2 + edge_y ** 2)

# Apply a threshold to the edge strength
threshold = 50  # Adjust this threshold as needed
edge_strength[edge_strength < threshold] = 0  # Remove weak edges

# Display the results
cv2.imshow('Edge Detection (x filter)', edge_x)
cv2.imshow('Edge Detection (y filter)', edge_y)
cv2.imshow('Edge Detection (Combined)', edge_strength)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
